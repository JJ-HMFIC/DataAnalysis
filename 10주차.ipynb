### 시계열 데이터


- 시간에 대해 모인 자료, 다른 데이터에 비해서 시계열적인 인사이트가 있다.
- datetime모듈 
- pandas timedelta , Timestamp 

> datetime은?
- date, time 연, 월 ,일 특정시각 등 나눠질 수 있는 형태


> Timestamp?
- 단일 객체에서 날짜 와 시간 모두 갖고 있다.

> timedelta
- 날짜에 대한 덧셈과 뺄셈 등 계산할 때 유용하다!

### 시계열 데이터를 만들기 위해서는
- 데이터프레임에서 시계열 형태의 컬럼을 찾고, 그 컬럼을 꼭 datetime으로 바꿔줘야 한다!

```python
crime.info()
```

```python
crime['REPORTED_DATE'] #문자열 형태로 있음
```

```python
crime['REPORTED_DATE'] = pd.to_datetime(crime['REPORTED_DATE'])
```

```python
crime.info()
# 3 REPORTED_DATE Dtype가 datetime으로 변환되어 있음 
```

### 해당 컬럼을 datetime으로 변경 후! 
- 인덱스 기준 설정! 인덱스를 시계열데이터로 만들자!

```python
crime_df = crime.set_index('REPORTED_DATE') #시계열 컬럼을 인덱스로 사용!
```

```python
crime_df
```

```python
#시계열 데이터를 인덱스로 변경한 후에는 슬라이싱, 원하는 요일, 연도 등 자유롭게 추출할 수 있다.
crime_df.loc['2014']
```

```python
crime_df.loc['2014-02-01']
```

```python
crime_df.loc['Dec 2014']
```

```python
crime_df.loc['Dec 2014'].sort_index()
```

```python
crime_df.loc['2014-12-01 17:42:00':'2014-12-25 22:36:00'].sort_index()
```

### Datetime index만 작동하는 메서드!

```python
crime_df.between_time('10:00','18:00').sort_index()
```

```python
## 내가 원하시는 시간도 추출 할 수 있음

crime_df.at_time('17:53').sort_index()
```
