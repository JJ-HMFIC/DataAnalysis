```python
df1 = pd.DataFrame({'A':[1,2,3],
                   'B':[4,5,6]})
df2 = pd.DataFrame({'A':[1,2,3],
                   'C':[7,8,9]})
```

```python
display(df1)
display(df2)
```

```python
print(df1)
# print와 display의 차이 
```

```python
#병합 진행

pd.concat([df1,df2],axis=0)
# axis = 축 
# axis= 0, 열을 중심으로 병합함 ↓
```

```python
pd.concat([df1,df2],axis=1)
# axis= 1,  행을 중심으로 병합함 →
```

```python
display(df1)
display(df2)
```

```python
pd.merge(df1, df2, how='left', on='A')
#on 교집합의 개념 , A를 중심으로 묶는다 
# how left, right, inner, outer ?
```

### 교집합이 2개 이상인 경우!

```python
df1 = pd.DataFrame({'A':[1,2,3],
                   'B':[4,5,6],
                   'C':[13,14,15]})
df2 = pd.DataFrame({'A':[1,2,3],
                   'B':[4,5,7],
                   'D':[10,11,12]})
```

```python
display(df1)
display(df2)
```

```python
pd.merge(df1,df2,how='inner',on=['A','B'])
#정말 교집합만 반환 A, B를 기준으로 모두 다 같아야 한다!
```

```python
df1
df2 # <= df2만 출력 
```

### 결측치 찾아보기!
- 누락된 값, 비어있는 값 의미

```python
import numpy as np
```

```python
df = pd.DataFrame({'A':[1,2,3,4,5,np.nan],
                  'B':[7,8,np.nan,9,10,11]})
```

```python
df
```

```python
df['A']+1
```

- NaN 값은 0의 개념이 아니라 어떤 값을 더하거나, 곱해도, 나눠도 다 NaN값이다.

```python
df.isna()#na값을 찾는다.
# nan 이면 True
# 아니라면 False 반환 
```

```python
df.isna().sum()
# nan개수 의 합 
```

```python
df.dropna() 
#전체 다 결측치 제거하는 법
```

```python
df
# 그러나 다른 함수와 마찬가리로 원본에 영향을 미치지 않는다. 
```

```python
#내가 원하는 컬럼에 결측치 제거 
df.dropna(subset = ['B'])
#B에 대한 결측치만 제거한다.
#제거한 인덱스를 스킵하여 나타냄 -> A도 영향을 받음 
# 인덱스 2가 출력되지 않음을 볼 수 있음 
```

```python
exam.loc[[3,7,12],['science']] = np.nan
# loc[[행],[열]] = 변수
# 행에 있는 값을 바꿀 때 쓴다
# 여기서 행 = 인덱스 번호 
```

```python
exam
# 원본의 값을 직접 바꾸기에 원본에 영향 o 
```

### 평균이 아닌 0으로 채운다

```python
exam['science'] = exam['science'].fillna((4+5)/3) #fillna na값을 채운다. #안에 숫자가 나오는 공식도 사용할 수 있다.
```

```python
exam
# 바뀌지 않음 
# 그 이유는 이미 fillna(61)로 인해 na 값이 61로 채워져
# 바꿀 na가 없기때문 
```

```python
exam.loc[[3,7,12],['science']] = np.nan
```

```python
exam['science'] = exam['science'].fillna((2/3)) 
# 다시 트라이 
```

```python
exam
# 0이 아닌 수식을 fillna의 인자로 사용 가능 
```
