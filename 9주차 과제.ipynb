## 필수과제 
### 기상청 데이터에서 
- 날씨 데이터를 받아오셔서 melt 함수를 이용해서 깔끔하게 정리해 주세요!
- 22년 10, 11월 / 21년 10월 11월 두 개를 비교해서 
- 날씨에 대한 비교도 간단하게 부탁드릴게요!

```python
import pandas as pd
```

```python
df2022 = pd.read_csv("2022년 날씨.csv", encoding = 'euc-kr')
```

```python
df2022
```

```python
df2022.head(10)
```

```python
df2021 = pd.read_csv("2021년 날씨.csv")
```

```python
df2021
```

```python
df2021.head(10)
```

```python
pd.melt(df2022, id_vars='일시')
```

```python
pd.melt(df2022, id_vars = ['지점번호','지점명']) 
```

```python
melt2022 = pd.melt(df2022, id_vars = ['지점번호','지점명','일시']) 
melt2022
```

```python
melt2022.groupby('variable').max()
```

```python
melt2022.groupby('variable').min()
```

```python
melt2021 = pd.melt(df2021, id_vars = ['지점번호','지점명','일시']) 
melt2021
```

```python
melt2021.groupby('variable').max()
```

```python
melt2021.groupby('variable').min()
```

```python
display(melt2021.groupby('variable').max())
melt2022.groupby('variable').max()
```

```python
display(melt2021.groupby('variable').min())
melt2022.groupby('variable').min()
```
