### 집단별로 요약하는 Groupby
- agg 더해서 통계치를 구하는 경우가 자주 발생!

- agg는 요약통계를 다 사용할 수 있다.
- mean, max, sum, min, std, median 등등

```python
exam.groupby('nclass')
```

```python
exam.groupby('nclass').mean()
# nclass 별로 묶어서 각 반 자료의 평균을 보여줌
# 같은 값을 하나로 묶어 통계 또는 집계 결과를 얻기 위해 사용하는 것이 groupby
```

```python
exam.groupby('nclass').agg(mean_math = ('math','mean'))
# 데이터프레임.groupby([컬럼1,컬럼2]).agg({계산하고싶은컬럼명1 : 집계함수 , 계산하고싶은컬럼명2 : 집계함수})
# 여기서는 mean_math를 정의하는데 
# mean_math란 math의 mean을 보여줌 
```

```python
exam.groupby('nclass').agg(mean_math = ('math','mean'), #평균
                          sum_math =('math','sum'), #합
                          median_math=('math','median'), #중앙값
                          ct = ('math','count'))#빈도,개수
```
