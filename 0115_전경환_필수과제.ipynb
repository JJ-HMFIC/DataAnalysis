#  과제1 
- 정규식에서 *, '+' 구글링해서 공유 주세요!

문자 또는 숫자가 0개 이상 나타남을 의미합니다. 예를 들어, a*는 0개 이상의 문자와 일치합니다.

## 오토마타의 star closure와 같음

```python
import re
```

```python
re.match('ab*c','ac')  # b가0개일 경우 
```

```python
re.match('ab*c','abc')
```

```python
re.match('ab*c','abbbbbbbbbc')
```

문자가 1개 이상 나타남을 의미합니다. 예를 들어, [nop]+는 n, o, p 글자가 하나 이상 있는 경우 일치합니다.

## 오토마타의 + 와 같은 의미

```python
re.match('ab+c','ac')  
```

```python
re.match('ab+c','abc')
```

```python
re.match('ab+c','abbbbbbbbbc') 
```

```python
re.search('ab+c','abbbbbbbbbc abbbc ac') 
```

### 과제2 약어를 이용해서 코드를 3개 이상씩 만들어 주세요!
- \d 모든 숫자
- \D 숫자를 제외한 모든 문자
- \w 영문, 대소문자, 숫자, 밑줄문자 
- \W 영문, 대소문자, 숫자, 밑물문자를 제외한 모든 문자

### \d

0123456789 범위의 모든 한 자리 숫자와 일치합니다.

실제 숫자 0-9는 숫자를 써서 표현할 수도 있습니다.

```python
re.match('\d','0123456789') # 처음만 선택 
```

```python
re.match('\d+','0123456789') 
```

```python
re.match('\d*','0123456789')
```

### \D
- 숫자를 제외한 모든 문자

```python
re.search('\D+','crasheso123@gmail.com')
```

```python
re.search('\D*','123@gmail.com')
```

```python
re.search('\D+','123@gmail.com')
```

### \w 
- 영문, 대소문자, 숫자, 밑줄문자 
- 
- 모든 글자, 숫자, 밑줄(_)과 일치합니다.

- 대소문자를 구별하는 실제 문자는 a-z 또는 A-Z의 문자를 입력하여 표현할 수도 있습니다.

```python
re.search('\w+','crasheso@gmail.com')
```

```python
re.search('\w+','2023/01/08')
```

### \W
- 영문, 대소문자, 숫자, 밑물문자를 제외한 모든 문자
- 문자 또는 숫자가 아닌 문자


```python
re.search('\W+','1+2+3+4=10') 
```

```python
re.search('\W*','1+2+3+4=10') 
```

```python
re.search('\W','010-1234-5678') 
```
