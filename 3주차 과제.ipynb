```python
mpg
```

```python
mpg.groupby('manufacturer').agg(mean_displ=('displ','mean'))
# 제조사별 배기량 평균 
```

```python
mpg.groupby('manufacturer')['displ','cty','hwy'].mean()
# 제조사별 배기량 도시연비 고속도로연비 
```

```python
mpg.query("manufacturer == 'audi'").groupby('model')['cty','hwy'].mean()
# 아우디 제조사의 모델별 도시연비,고속도로 연비 평균 
# query의 인자는 '조건문' -> ==
```

```python
prob1= mpg.groupby('manufacturer').agg(mean_displ=('displ','mean'))
prob1.reset_index()
# 인덱스를 컬럼으로 바꾸기 
```

```python
mpg.groupby('manufacturer',as_index=False).agg(mean_displ=('displ','mean'))
# 다른 방식 , 변수 선언 없이 
```

```python
mpg.groupby('trans')['displ','cty','hwy'].mean()
# 구동방식의 평균통계 
```

```python
mpg.groupby(['model','drv']).agg(mean_hwy=('hwy','mean')).sort_values('mean_hwy',ascending=False)
# 고속도로 연비 평균이 높은순으로 정렬하고 
# 모델과 구동방식 별 고속도로 연비 평균 
```
