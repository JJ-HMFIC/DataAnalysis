### displ

```python
mpg.sort_values('displ').head()
# 배기량 오름차순 정렬 후 상위 5개 출력 
```

```python
mpg.groupby('manufacturer').agg(mean_displ=('displ','mean'))
# 제조사별 배기량 평균 
```

```python
mpg.groupby('model').agg(mean_displ=('displ','mean')).sort_values('mean_displ').head()
# 모델별 배기량 평균 중 상위 다섯개 출력 
```

```python
mpg.groupby('model').agg(mean_displ=('displ','mean')).sort_values('mean_displ').tail()
# 하위 다섯개 출력 
```

```python
mpg.groupby('model').agg(mean_displ=('displ','mean')).sort_values('mean_displ').query('mean_displ>3.5')
# 모델 별 평균 배기량 중 3.5 이상 출력 
```

### year

```python
mpg['year'].value_counts()
# 연도별 판매량 
```

```python
mpg.query('year==1999 and displ<3.0 and category !="compact"').sort_values('displ')
# 1999년도,배기량 <3.0 그리고 compact 등급이 아니면서 배기량으로 정렬함
```

```python
mpg_year_tmp = mpg.query('year==1999 and displ<3.0 and category !="compact"').sort_values('displ')
mpg_year_tmp.query('cty>20 and hwy>20')
# 위의 자료에서 도시 연비와 고속도로 연비가 20보다 높은 자동차를 출력 
```

```python
mpg_year_tmp.query('cty>20 and hwy>20').value_counts('category')
# 위의 조건에서 category는 다음과 같다 
```

```python
mpg_year_tmp.groupby('model').agg(mean_cty=('cty','mean'))
# 1999년도,배기량 <3.0 그리고 compact 등급이 아니면서 배기량으로 정렬한 자료중에
# 모델별 평균 배기량 출력 
```

### cyl

```python
mpg['cyl'].value_counts()
# 실린더 개수 카운팅
```

```python
mpg.query('cyl==8 and manufacturer =="honda"')
# 혼다 제조사의 실린더 개수가 8개인 자동차 출력
# 혼다는 8개로 만들지 않음을 알 수 있다.
```

```python
mpg.query('cyl==8').groupby('model').agg(mean_hwy=('hwy','mean'))
# 실린더가 8개인 모델 별 고속도로 연비 평균 
```

```python
mpg.query('cyl==8').groupby('manufacturer').agg(mean_hwy=('hwy','mean'),mean_cty=('cty','mean'))
# 실린더가 8개인 제조사 별 고속도로 연비와 도시연비 평균 
```

```python
mpg.query('cyl==6 and displ<3.1')
#실린더 개수가 6개면서 배기량이 3.1 미만인 자료 출력 
```

### trans

```python
mpg['trans'].value_counts()
# 변속기 종류
```

```python
mpg_tmp_trans = mpg[['model','displ','cty','hwy','trans']]
mpg_tmp_trans.query('trans =="auto(l5)" and displ<2.5')
#변속기 종류가 위와 같고 배기량이 2.5미만인 자료들 출력 
```

```python
mpg_tmp_trans.query('trans =="auto(l5)"').groupby('model').agg(mean_hwy=('hwy','mean'),mean_cty=('cty','mean'))
# 변속기 종류가 위와 같을 때 모델별 평균 고속도로 연비와 평균 도시 연비 
```

```python
mpg['drv'].value_counts()
```

```python
mpg.groupby('drv').agg(mean_hwy=('hwy','mean'),mean_cty=('cty','mean'))
# 구동방식 별 연비 측정 
```
