### 누락값을 실제로 처리해 보자!

- na값을 채우는 방법 중 
1. fillna()

```python
ebola.iloc[0:10, 0:5]
```

```python
ebola.iloc[0:10, 0:5]['Cases_Liberia'].mean()
# 0을 넣지 않았을 때!
```

```python
ebola.fillna(0).iloc[0:10, 0:5]['Cases_Liberia'].mean()
# 0으로 대체했을 때!
```

### 다른 방법으로 누락값을 대체하는 법
- fillna(method =' ')

- ffill - 누락값이 나타나기 전의 값으로 누락값을 변경
- bfill - 누락값이 나타난 이후 첫 번째 값으로 앞쪽의 누락값을 변경한다.

```python
ebola.fillna(method='ffill').iloc[0:10, 0:5]['Cases_Liberia'].mean()

# ffill로 할 경우 결측값을 바로 위 값과 동일하게 변경합니다.
```

```python
ebola.fillna(method='bfill').iloc[0:10, 0:5]['Cases_Liberia'].mean()
# bfill로 할경우 결측값을 바로 아래 값과 동일하게 변경합니다.
```

```python
##평균으로 대체합니다.

ebola.iloc[0:10, 0:5]['Cases_Liberia'].mean()
```

```python
# 평균 값 8049.17
ebola.fillna(8049.17).iloc[0:10, 0:5]['Cases_Liberia'].mean()# 평균으로 대체 
```

### 누락값을 대체하는 법
- 보간법 
- interpolate(method='')

- 선형보관, spline 보간법 

```python
ebola['Cases_Liberia'].plot()
# 현재 결측치를 포함하여 나타낸 그래프
```

```python
ebola.fillna(8049.17)['Cases_Liberia'].plot()
# 결측치 제외한 평균으로 채운 후 그래프 
```

```python
ebola.interpolate(method='linear')['Cases_Liberia'].plot()
# 선형 보간법 
"""
선형 보간법(Linear Interpolation)은 2개의 인접한 관측값이 주어졌을 때 
그 사이에 위치한 값을 추정하기 위해 임의적인 직선을 가정하고 선형적으로 계산하는 방법
"""
```

### 스플라인 보간법 
- 각 데이터의 포인트 구간별로 낮은 수준의 다항식 보간을 사용, 직선으로 연결되는 보간법 보다는 좀 더 부드럽게 연결이 되어서 다항식 항목을 선택한다.

```python
ebola.interpolate(method='quadratic')
```

```python
ebola.interpolate(method='quadratic')['Cases_Liberia'].plot()
#  quadratic(부드러운 보정)
```

```python
ebola.interpolate(method='linear')['Cases_Liberia'].plot()
```

```python
ebola.interpolate(method='dasd')
```

 method must be one of ['linear', 'time', 'index', 'values', 'nearest', 'zero', 'slinear', 'quadratic', 'cubic', 'barycentric', 'krogh', 'spline', 'polynomial', 'from_derivatives', 'piecewise_polynomial', 'pchip', 'akima', 'cubicspline']

### 열과 피벗
- 엑셀에서 피벗의 개념

- melt 메서드를 이용해서 깔끔하게 데이터럴 정리하는데 유용하다!

- melt의 메서드 인자

- id_vars : 위치를 그대로 유지할 열의 이름을 지정

- value_vars : 행으로 위치를 변경할 열의 이름을 지정

- var_name: value_vars로 위치 변경할 열의 이름 지정

- value_name : var_name의 위치로 변경할 열의 데이터를 저장한 열의 이름
